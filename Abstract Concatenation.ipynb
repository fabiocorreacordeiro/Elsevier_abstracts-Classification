Importando bibliotecas

```python
import pandas as pd
import json
```

Lendo os arquivos JSON com os textos extraídos utilizando a API da Elsevier. Para os aquivos provenientes de artigo cujo um dos autores é afiliado a uma empresa de Óleo & Gás foi acrescentado a coluna "O&G" com o valor 1 e uma coluna "Affiliate" com o nome da empresa. Para os demais artigos foi atribuído o valor 0 para "O&G" e "outros" para a coluna "Affiliate". 

```python
# Inicialmete lemos o arquivo referente aos textos da "Petrobras"
abstract_petrobras = pd.read_json(path_or_buf="Elsevier_abstracts - Petrobras.json")
abstract_petrobras['O&G'] = 1
abstract_petrobras['Affiliate'] = 'Petrobras'
abstract_petrobras
```

```python
# Lendo os arquivos da demais Majors
abstract = abstract_petrobras
companies = ["BP America",
             "BP Exploration",
             "BP International",
             "BP United Kingdom",
             "British Petroleum",
             "Chevron",
             "Eni",
             "Equinor",
             "Exxon",
             "Repsol",
             "Shell",
             "Statoil",
             "Total"]

for company in companies:
    path = "Elsevier_abstracts - " + company + ".json"
    abstract_company = pd.read_json(path_or_buf = path)
    abstract_company['O&G'] = 1
    abstract_company['Affiliate'] = company
    abstract = pd.concat([abstract, abstract_company])
    
abstract
```

```python
# Lendo os arquivos de autores não filiados a empresas de O&G
years = ["2018", "2017", "2016", "2015", "2014", "2013", "2012", "2011", "2010",
         "2009", "2008", "2007", "2006", "2005", "2004", "2003", "2002", "2001", "2000"]

for year in years:
    path = "Elsevier_abstracts - " + year + ".json"
    abstract_year = pd.read_json(path_or_buf = path)
    abstract_year['O&G'] = 0
    abstract_year['Affiliate'] = "outros"
    abstract = pd.concat([abstract, abstract_year])
    
abstract = abstract.reset_index()
abstract = abstract.drop(columns=['index'])
abstract["Date"] = abstract["Date"].astype(str)
abstract
```

Salvando arquivo consolidado

```python
abstract.to_json("Elsevier_abstract - Consolidado.json")
```
